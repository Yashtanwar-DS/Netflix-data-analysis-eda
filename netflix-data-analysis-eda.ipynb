```python
df["Type"].value_counts()
df["Type"].value_counts().plot(kind = "bar")
```

```python
df["Country"].value_counts().head(10)
```

```python
df["Country"].value_counts().head(10).plot(kind = "pie",autopct = ("%0.1f%%"),labeldistance = 1.1,startangle = 90)
```

```python
df["Genre"].value_counts().head(5)
```

```python
df["Genre"].value_counts().head(5).plot(kind = "bar",edgecolor = "r")
```

#Combination Analysis
